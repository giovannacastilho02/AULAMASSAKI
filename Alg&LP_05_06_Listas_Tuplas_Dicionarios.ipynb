# **Coleções no Python: Listas, Tuplas e Dicionários**


# **Listas**
Lista é uma coleção de **valores indexados**, em que cada valor é identificado por um índice. O primeiro item na lista está no índice 0, o segundo no índice 1 e assim por diante.

# **Tuplas**
Tupla é uma estrutura de dados semelhante a lista. Porém, ela tem a característica de ser **imutável**, ou seja, após uma tupla ser criada, ela não pode ser alterada.

# **Dicionários**
Os dicionários representam coleções de dados que contém na sua estrutura um conjunto de **pares chave/valor**, nos quais cada chave individual tem um valor associado. Esse objeto representa a ideia de um mapa, que entendemos como uma coleção associativa desordenada. A associação nos dicionários é feita por meio de uma chave que faz referência a um valor.

```python
professores = ['Carla', 'Daniel', 'Ingrid', 'Roberto']
estacoes = ('Primavera', 'Verão', 'Outono', 'Inverno')
cliente = {
    'Nome': 'Fábio Garcia',
    'email' : 'fabio_garcia_9@outlook.com'
}

print(type(professores)) # list
print(type(estacoes)) # tuple
print(type(cliente)) # dict
```

## **APESAR DAS TUPLAS SEREM IMUTÁVEIS, EU POSSO**
Acrescentar elementos a uma tupla Usando o método append()
Um novo elemento é adicionado ao final da lista usando o método append().

```python
cliente1 = ('Massaki', '123.456.789-10', '(11)99999-9999')
clientes = []
clientes.append(cliente1)
print(clientes)
```

```python
cliente2 = ('Zé', '112.213.141-15', '(11)99123-4567')
clientes.append(cliente2)
print(clientes)
```

## **O que é um DataFrame?**
Um Pandas DataFrame é uma estrutura de dados bidimensional, como uma matriz bidimensional ou uma tabela com linhas e colunas.

```python
import pandas as pd
df = pd.DataFrame(clientes)
df
```

# **O método concat do pandas**
## Imagine que você tem várias tabelas de dados (ou DataFrames, como são chamados no pandas ) e você precisa juntá-los em uma única tabela. É aqui que o método concat entra em jogo. O concat permite que você “cole” (concatene) DataFrames ao longo de um eixo específico, seja vertical ou horizontalmente. Dessa forma, você consegue unir DataFrames um abaixo do outro, ou um ao lado do outro.

### **Para utilizar o pd.concat, você precisa passar uma lista de DataFrames que deseja concatenar e, opcionalmente, pode especificar o eixo. Por padrão, o concat une os DataFrames verticalmente, empilhando um sobre o outro. Vamos ver um exemplo prático:**

```python
import pandas as pd

df1 = pd.DataFrame({"NOME": ['João', 'Maria'],
                    "CPF": ['123.456.789-10', '112.213.141-15'],
                    "TEL": ['454545454', '(11)99123-4567']})
# Use pd.concat to combine DataFrames
df2 = pd.concat([df, df1], ignore_index=True)
df2
```
